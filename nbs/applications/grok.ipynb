## Multi Agent + Web Search

In this example we combine Grok's multi-agent capabilities with web search to find the best hotel for our ski trip.

NOTE: Multi-agent can get expensive. The call below costs me ~$0.55, even with low reasoning effort.

```python
c = Chat("x-ai/grok-4.20-multi-agent:online")
```

```python
r = c(
    "What is the best hotel I can book in Sierra Nevada for 15-19 april under 1000 euros and as close to the ski lifts as possible?",
    extra_body={"reasoning": {"effort": "low"}},
)
```

```python
print(r)
```
